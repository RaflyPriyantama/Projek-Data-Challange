```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import Sort, reviews
result, continuation_token = reviews(
    'com.shopee.id',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=1000000,
    filter_score_with=None
)
```

```python
df_scrap = pd.DataFrame(np.array(result),columns=['review'])
df_scrap = df_scrap.join(pd.DataFrame(df_scrap.pop('review').tolist()))
df_scrap.head()
```

```python
my_df = df_scrap[['userName', 'score', 'at', 'content']]
```

```python
df_cod = my_df[my_df['content'].str.contains('cod', case=False, na=False)]

pd.set_option('display.max_rows', None)
df_cod
```
