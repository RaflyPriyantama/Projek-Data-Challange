```python
import pandas as pd
import re

# 1. Baca data
df = pd.read_csv("scrapped_data_cod.csv")
df.columns

# 2. Hapus kolom username (ganti nama kolom sesuai nama aslinya di dataset)
if 'userName' in df.columns:
    df = df.drop(columns=['userName'])

# 3. Hapus duplikat baris
df = df.drop_duplicates()

# 4. Cleaning teks (hapus tanda baca, huruf kecil, hapus link, emoji, spasi berlebih)
def clean_text(text):
    text = str(text).lower()  # ubah ke huruf kecil
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)  # hapus tanda baca dan simbol
    text = re.sub(r"\s+", " ", text).strip()  # hapus spasi berlebih
    return text

# Terapkan ke semua kolom bertipe object (teks)
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].apply(clean_text)

# 6. HAPUS BARIS YANG MENGANDUNG ANGKA DI KOLOM CONTENT
if 'content' in df.columns:
    # Simpan data sebelum dihapus untuk backup
    backup_df = df.copy()
    
    # Filter baris yang mengandung angka di kolom content
    rows_with_numbers = df[df['content'].str.contains(r'\d', na=False)]
    
    print("STATISTIK SEBELUM PENGHAPUSAN:")
    print(f"Jumlah baris dengan angka: {len(rows_with_numbers)}")
    print(f"Total semua baris: {len(df)}")
    
    # Hapus baris yang mengandung angka di content
    df = df[~df['content'].str.contains(r'\d', na=False)]
    
    print("\nSTATISTIK SETELAH PENGHAPUSAN:")
    print(f"Baris yang dihapus: {len(rows_with_numbers)}")
    print(f"Sisa baris: {len(df)}")
    print(f"Pengurangan: {len(backup_df) - len(df)} baris")
    
    # Simpan data yang dihapus ke file terpisah untuk referensi
    rows_with_numbers.to_csv("data_yang_dihapus.csv", index=False)
    print(f"\nData yang dihapus disimpan ke: data_yang_dihapus.csv")
    
else:
    print("Kolom 'content' tidak ditemukan dalam dataset")

# 7. Simpan hasil akhir
df.to_csv("scrapped_data_cod_cleaned.csv", index=False)
print(f"\nData bersih disimpan ke: scrapped_data_cod1_cleaneddd.csv")
print(f"Final: {len(df)} baris data")
```
